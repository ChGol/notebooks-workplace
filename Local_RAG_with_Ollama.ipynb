```python
# Setup local RAG using using LLAMA3 via https://github.com/ollama/ollama-python and wikipedia documents
from llama_index.llms.ollama import Ollama
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.readers.wikipedia import WikipediaReader
from llama_index.core import VectorStoreIndex

# Setup high timeout for Colab GPU
llama = Ollama(model="llama3", request_timeout=600.0)
```

```python
embed = HuggingFaceEmbedding(
    model_name="hkunlp/instructor-base"
)

Settings.llm = llama
Settings.chunk_size = 512
Settings.chunk_overlap = 75
Settings.embed_model = embed
```

```python
documents = WikipediaReader().load_data(["History of science fiction films"])
vector_index = VectorStoreIndex.from_documents(documents)
```

```python
query_engine = vector_index.as_query_engine(similarity_top_k=3)
response = query_engine.query("What is the oldest scifi movie?")
print(response)
```

```python
# Graph RAG with LLAMA
from llama_index.core import KnowledgeGraphIndex
from llama_index.core.graph_stores import SimpleGraphStore
from llama_index.core import StorageContext
from llama_index.readers.huggingface_fs import HuggingFaceFSReader
from pyvis.network import Network
```

```python
loader = HuggingFaceFSReader()
graph_documents = loader.load_data(
    "datasets/SauravMaheshkar/threads-ask-ubuntu")

# Setup the storage context
graph_store = SimpleGraphStore()
storage_context = StorageContext.from_defaults(graph_store=graph_store)

#Construct the Knowlege Graph Undex
vector_index_with_graph = KnowledgeGraphIndex.from_documents(
                                           documents=graph_documents,
                                           max_triplets_per_chunk=3,
                                           storage_context=storage_context,
                                           embed_model=embed,
                                           include_embeddings=True)
```

```python
graph_query_engine = vector_index_with_graph.as_query_engine(include_text=True,
                                     response_mode ="tree_summarize",
                                     embedding_mode="hybrid",
                                     similarity_top_k=3,)
response = graph_query_engine.query("How to install CUDA?")
print(response)
```
